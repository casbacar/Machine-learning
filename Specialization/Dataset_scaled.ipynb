## 2. Scaling the data

```python
# We are going to exclude the columns DATE and Month from scaling
# Columns to exclude
exclude_cols = df[['DATE', 'MONTH']]

# Separate features to scale and not to scale
features_to_scale = [col for col in df.columns if col not in exclude_cols]

# Scale only selected features
scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[features_to_scale] = scaler.fit_transform(df[features_to_scale])
```

```python
# We create the scaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df.drop(['DATE', 'MONTH'], axis=1))
```

```python
# We convert the numpy array generated in the previous step into a dataframe

scaled_df = pd.DataFrame(df_scaled, columns=df.columns, index=df.index)
```
